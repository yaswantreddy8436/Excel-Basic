1. Difference Between Relative, Absolute, and Mixed Cell Referencing

Relative Referencing: This is the default. When you copy a formula, the cell references change based on the relative position of rows and columns. (Example: A1)
Absolute Referencing: The cell reference stays fixed even if the formula is copied elsewhere. It is denoted by dollar signs. (Example: $A$1)
Mixed Referencing: Only part of the reference is fixed—either the row or the column. (Example: $A1 fixes the column; A$1 fixes the row)
